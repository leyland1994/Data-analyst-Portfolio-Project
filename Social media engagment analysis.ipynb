# Social Media Analysis and Visualization


```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random

```

# Generating random data for the social media data

```python
# Define categories and size
categories = ['Food', 'Travel', 'Fashion', 'Fitness', 'Music', 'Culture', 'Family', 'Health']
n = 500  # Number of records

# Create random data
data = {
    'Date': pd.date_range('2021-01-01', periods=n),
    'Category': [random.choice(categories) for _ in range(n)],
    'Likes': np.random.randint(0, 10000, size=n)
}
```

# Load the data into a Pandas DataFrame and Explore the data

```python
# Create DataFrame
df = pd.DataFrame(data)
df.head()  # Display the first few rows
```

```python
# Check the structure of the DataFrame
print(df.info())
```

```python
# Get summary statistics
print(df.describe())
```

```python
# Count of each category
print(df['Category'].value_counts())
```

 # Clean the Data

```python
# Check for null values
print("Null values in the DataFrame:")
print(df.isnull().sum())

# Drop null values and duplicates
df = df.dropna().drop_duplicates()

# Ensure correct data types
df['Date'] = pd.to_datetime(df['Date'])
df['Likes'] = df['Likes'].astype(int)
```

# Visualize the Data

```python
# Histogram of Likes
# sns.histplot(df['Likes'], bins=20, kde=True)
# plt.title('Distribution of Likes')
# plt.xlabel('Likes')
# plt.ylabel('Frequency')
# plt.show()

sns.distplot(df['Likes'], bins=20, kde=True, hist=True)
plt.title('Distribution of Likes')
plt.xlabel('Likes')
plt.ylabel('Frequency')
plt.show()

# Boxplot of Likes by Category
sns.boxplot(x='Category', y='Likes', data=df)
plt.title('Likes by Category')
plt.xticks(rotation=45)
plt.show()
```

 # Analyze the Data

```python
# Overall mean of Likes
mean_likes = df['Likes'].mean()
print(f"Mean Likes: {mean_likes}")

# Mean Likes per category
category_means = df.groupby('Category')['Likes'].mean()
print("Mean Likes by Category:")
print(category_means)
```

1. Histogram of Likes

The histogram shows how the number of likes is distributed across all posts.
From the shape of the distribution, we can see that:

Most posts receive moderate numbers of likes, forming a cluster in the lower to mid-range.

As the number of likes increases, the frequency of posts decreases sharply.

Only a small number of posts achieve very high engagement, which creates a long tail on the right.

This tells us that viral posts are rare, and the majority of content performs within a normal range  a common trend on social media. It also suggests that the overall audience engagement is somewhat concentrated and predictable, with only a few outliers standing out.

2. Boxplot of Likes by Category

The boxplot gives a deeper view of how likes vary within each category, not just the average.

Heres what the boxplot reveals:

The Lifestyle and Fitness categories show higher median likes, meaning posts in these categories generally perform well even before considering outliers.

The spread (the height of the box) shows how consistent engagement is. A taller box means the category has a wide range of performance.

Some categories show significant outliers, which appear as dots outside the box  these represent posts that received extremely high likes compared to others in the same category.

Categories with a narrow box and few outliers have more stable engagement; categories with a wide box and many outliers have unpredictable performance.

In short, the boxplot helps you see both the consistency and the extremes for each category. This adds depth beyond simple averages and shows which categories are reliably engaging and which ones have more variable results.
